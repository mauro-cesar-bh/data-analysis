## Context
A Pix Invoice is one of the main methods for an enterprise to charge customers. Similarly to a Boleto, an Invoice is first issued with an expected amount to be paid, and then customers pay this Invoice using their bank app, either by scanning a QR Code or copying the payment information.

## Task 1 - Low Activity Periods

- **When** is it best to deploy our Invoice application?

## Solution
The best time to deploy **is between 2:00 and 5:00 am**, if possible, on Sundays.

## Task 2 - Payment Conversion
Our customers are interested in having analytical data on payment conversion.
- 2.1 What percentage of Invoices are **paid** and what percentage of Invoices are **paid after the due date**? 
    - Paid = Paid before due date + Paid after due date
- 2.2 How has the overdue **payment rate evolved** in the available months? 
    - Overdue payment rate = Invoices paid after due date / Total Invoices
- 2.3 Which **factors** influence whether the Invoice will be paid in time?

## Solution

78.9% of invoices are paid, and 6.6% are paid after the due date

### 2.2 How has the overdue **payment rate evolved** in the available months? 
The overdue Payment Rate is increasing over the months.

## Task 3 - Reversed Invoices
Invoices may be reversed partially or totally (reverse amount = $0).

3.1 What percentage of Invoices are **partially or totally reversed**?

3.2 How **long** does it usually take for an Invoice to be reversed (reversed log time - invoice created time)?

3.3 Which **factors** influence whether the Invoice will be reversed and whether the reversal is partial or total?

## Solution 
### 3.1 What percentage of Invoices are **partially or totally reversed**?
Only 18bps (0.18% of the invoices were reversed)

### 3.2 How **long** does it usually take for an Invoice to be reversed?
It usually take 39 minutes (median), but due skewness it has a 297 minutes on average.

### 3.3 a) What factors influences to be **reverted**
The main factors are: 
- Amount of the invoice: Lower amounts tends to have higher chances of being reverted
- Weekday: End of week day tends to have lower chances of being reverted, with Friday (4) being only 14 bps
